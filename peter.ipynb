```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
url = "https://www.betus.com.pa/sportsbook/lines/"
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}
response = requests.get(url, headers=headers)
```

```python

session = requests.Session()
response = requests.get(url)


# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content of the page
    soup = BeautifulSoup(response.text, 'html.parser')

    # Extract the data you need using Beautiful Soup methods
    # For example, let's extract all the text from paragraph (p) tags
    table = soup.find_all('table')

    # Print the extracted text
    for tables in table:
        print(tables.text)
else:
    print(f"Error: Unable to retrieve the page. Status code: {response.status_code}")
```

```python
table = soup.find("table")
all_values = []
# Iterate through rows and columns
for row in table.find_all('tr'):
    row_values = []
    for cell in row.find_all(['th', 'td']):
        row_values.append(cell.text.strip())
    all_values.append(row_values)
# Print or use the extracted values
for row in all_values:
    print(row)   
```
